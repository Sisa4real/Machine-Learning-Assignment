# Exploring Visualizations

```python
#Importing the reqiured ibraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
 
```

```python
#Loading the datasets
pumpkins_data = pd.read_csv('US-pumpkins.csv')
```

```python
#Viewing the first few rows
pumpkins_data.head()
```

```python
pumpkins_data.isnull().sum()
```

```python
pumpkins_data.describe()
```

```python
pumpkins_data.info
```

```python
# using seaborn
sns.scatterplot(data=pumpkins_data,x='High Price', y= 'City Name')
plt.title('Relationship between City Name and High Prices')
plt.xlabel('High Price')
plt.ylabel('City Name')
plt.show()
```

```python
# using matplotlib
plt.scatter('Low Price','High Price', data=pumpkins_data)
plt.title('Relationship between High and Low Prices')
plt.xlabel('Low Price')
plt.ylabel('High Price')
plt.show()
```
